```python
r = 0.8
s = -0.8
f = 0.1
alpha = 0.045
beta = 0.03
```

```python
def lotka_volterra(P, t=0):
    return np.array(
        [(r-f)*P[0]- alpha*P[0]*P[1],
         (s-f)*P[1]+ beta*P[0]*P[1]]
    )
    
```

```python
from scipy.integrate import odeint
```

```python
t = linspace(0, 60, 1000)
```

```python
P0 = np.array([10, 5])
```

```python
P = odeint(lotka_volterra, P0, t)
```

```python
fish, shark = P.T
```

```python
plot(t, fish, label ='fish')
plot(t, shark, label ='shark')
plt.legend()
```

```python
plot(fish, shark)
plt.xlabel('fish')
plt.ylabel('shark')
```

```python
plot(fish, shark)
plt.xlabel('fish')
plt.ylabel('shark')
ss_fish = (f-s)/beta
ss_shark = (r-f)/alpha
plt.axvline(x=ss_fish, ls='--')
plt.axhline(y=ss_shark, ls='--')
```
